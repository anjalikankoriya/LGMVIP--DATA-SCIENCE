```python
#import library
from google.colab.patches import cv2_imshow
import cv2
```

```python
#loading the image
image = cv2.imread("dog.jpg")
cv2_imshow(image)
```

```python
#converting to gray scale image
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
#Tansforming into negative image
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

```python
#Making image blur using gaussian function
blur = cv2.GaussianBlur(img_invert, (21, 21), 0)
```

```python
#inverting the blurred image
blur_invert = 255-blur
```

```python
#sketch
sketch = cv2.divide(img_gray, blur_invert, scale=248.0)
cv2_imshow(sketch)
```
